```python
import annual
import pandas as pd
```

```python
consumption_df = pd.DataFrame({'Year': annual.energy_consumption[:, 0], 
                   'Natural gas': annual.energy_consumption[:, 1]/1000, 
                   'Electricity': annual.energy_consumption[:, 2]/1000})
```

```python
consumption_df.plot(x='Year', y=['Natural gas', 'Electricity'], figsize=(12,6), grid=True, ylabel="Annual energy consumption [TWh]")
```

```python
generation_df = pd.DataFrame({'Year': annual.wind_generation[:, 0], 
                   'Wind energy': annual.wind_generation[:, 2]/1000})
```

```python
generation_df.plot(x='Year', y='Wind energy', figsize=(12,6), grid=True, ylabel="Annual energy generation [TWh]")
```

```python
generation_df['increase'] = generation_df['Wind energy'].diff()
```

```python
generation_df.plot(x='Year', y='increase', figsize=(12,6), grid=True, ylabel="Annual energy generation increase[TWh]")
```
